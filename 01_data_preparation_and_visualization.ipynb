```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read data from a CSV file
data = np.genfromtxt('Rainfall_and_Streamflow_Dataset.csv', delimiter=',', skip_header=1)
# extract rainfall and streamflow data
#collective_rainfall = data[:, 0]  # assuming collective rainfall is in the first column
rainfall = data[:, 1]  # assuming rainfall is in the first column
streamflow = data[:, 2]  # assuming streamflow is in the second column
plt.figure(figsize=(10, 5))
# plot rainfall and streamflow data
plt.plot(rainfall, label='Rainfall', color='blue')
plt.plot(streamflow, label='Streamflow', color='green')
plt.title('Rainfall and Streamflow Data')
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
# Save the plot to a file
plt.savefig('rainfall_streamflow_plot.png', dpi=300)
plt.close()  # Close the plot to free up memory
```

```python
plt.figure(figsize=(10, 4))
plt.plot(data[:, 3], data[:, 1], label="Rainfall (mm)", color='blue')
plt.plot(data[:, 3], data[:, 2], label="Streamflow (m/s)", color='green')
plt.legend()
plt.title("Rainfall and Streamflow Over Time")
plt.xlabel("Date")
plt.ylabel("Value")
plt.grid(True)
plt.tight_layout()
plt.show()
plt.close()  # Close the plot to free up memory
```
